```python
# in this cell we check for each column if its empty. If it is, we fill categorical data with median value and  
# numerical ones with average value of the column

cols = list(train_data)

for col in cols:
    if (df[col].isnull().sum()) > 1:
        if isinstance(train_data[col][1], str) :
            train_data.fillna(train_data[col].median(), inplace = True)
        else: 
            train_data.fillna(train_data[col].median(), inplace = True)

cols = list(test_data)

for col in cols:
    if (df[col].isnull().sum()) > 1:
        if isinstance(test_data[col][1], str) :
            test_data.fillna(test_data[col].median(), inplace = True)
        else: 
            test_data.fillna(test_data[col].median(), inplace = True)           

# df = pd.DataFrame(test_data)
# df.head(20)

#2DO: see why at this step, some of the values in the test data are still "NaN" even though we apply 
# the same method of filling them in the test data
```

```python
# in this step, we will remove columns with values that are mostly zeros before converting catigorical data 
# into numerical ones. 

# first we check which columns have values that are mostly zeros

cols = list(train_data)


for col in cols:
#     print(col)
#     print(df[col].unique())
    if ( (df[col].astype(bool).sum(axis=0)/1460) < 0.40 ): # if the number of the non-zero value is less than 40%
        print(col)
        train_data = train_data.drop(columns = col)

cols = list(test_data)
for col in cols:
#     print(col)
#     print(df[col].unique())
    if ( (df[col].astype(bool).sum(axis=0)/1460) < 0.40 ): # if the number of the non-zero value is less than 40%
        print(col)
        test_data = test_data.drop(columns = col)
```

```python
# before this step fill in missing values.
# Now we turn catigorical values into numerical values


# train_data['MSZoning'] = train_data['MSZoning'].astype('category').cat.codes

cols = list(train_data)

for col in cols:
    if isinstance(train_data[col][1], str) :
        train_data[col] = train_data[col].astype('category').cat.codes
        
df = pd.DataFrame(train_data)
# df.head(20)

cols = list(test_data)
for col in cols:
    if isinstance(test_data[col][1], str) :
        test_data[col] = test_data[col].astype('category').cat.codes
        
df2 = pd.DataFrame(test_data)
# df2.head(20)
```
